```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("data/cleaned/ames_features.csv")
df.head()
```

```python
# Mean and standard deviation by hand with NumPy
target = df["SalePrice"].values

mean_value = np.sum(target) / len(target)
std_value = np.sqrt(np.sum((target - mean_value) ** 2) / len(target))

print("Manual mean:", round(mean_value, 2))
print("Manual std:", round(std_value, 2))
```

```python
# Standardize one column by hand
x = df["Gr_Liv_Area"].values
x_mean = np.mean(x)
x_std = np.std(x)

z_manual = (x - x_mean) / x_std

scaler = StandardScaler()
z_sklearn = scaler.fit_transform(df[["Gr_Liv_Area"]]).flatten()

comparison = pd.DataFrame({
    "manual": z_manual[:5],
    "sklearn": z_sklearn[:5]
})

comparison
```

```python
# Cosine similarity between the highest-price and lowest-price homes
numeric_df = df.select_dtypes(include=np.number)

high_row = numeric_df.loc[df["SalePrice"].idxmax()]
low_row = numeric_df.loc[df["SalePrice"].idxmin()]

# Remove the target itself so the comparison is more fair
high_vec = high_row.drop(labels=["SalePrice"]).values
low_vec = low_row.drop(labels=["SalePrice"]).values

cosine_similarity = np.dot(high_vec, low_vec) / (
    np.linalg.norm(high_vec) * np.linalg.norm(low_vec)
)

print("Cosine similarity:", round(float(cosine_similarity), 4))
```

```python
# Simple probability estimate
high_quality = df[df["Overall_Qual"] >= 8]
probability = (high_quality["SalePrice"] > 300000).mean()

print("Estimated probability:", round(float(probability), 4))
```

## Simple math summary
- I calculated the mean and standard deviation using NumPy.
- I standardized `Gr_Liv_Area` by hand and compared it with `StandardScaler`.
- I used cosine similarity to compare two very different homes.
- I estimated the probability that a high-quality home sells for more than 300,000.
