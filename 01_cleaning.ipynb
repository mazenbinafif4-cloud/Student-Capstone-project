```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("data/raw/AmesHousing.csv")
df.head()
```

```python
# Make column names easier to use
df.columns = [c.strip().replace(" ", "_").replace("/", "_") for c in df.columns]

print("Shape:", df.shape)
df.info()
```

## Fixing Data Types
- MS_SubClass represents categories, so I converted it to string
- Mo_Sold is a month, so I treated it as a category (string)

```python
# Fix column types (required: at least 2)
df["MS_SubClass"] = df["MS_SubClass"].astype(str)
df["Mo_Sold"] = df["Mo_Sold"].astype(str)

# Check
df[["MS_SubClass", "Mo_Sold"]].dtypes
```

```python
# Check missing values
missing = df.isnull().sum().sort_values(ascending=False)
missing[missing > 0].head(15)
```

```python
# Check duplicates
df.duplicated().sum()
```

```python
# Quick boxplot for the target
plt.figure(figsize=(8, 4))
plt.boxplot(df["SalePrice"])
plt.title("SalePrice Boxplot Before Cleaning")
plt.ylabel("SalePrice")
plt.show()
```

```python
def clean_data(df):
    df = df.copy()

    # Drop columns with too many missing values
    drop_cols = ["Pool_QC", "Misc_Feature", "Alley", "Fence", "Fireplace_Qu"]
    drop_cols = [c for c in drop_cols if c in df.columns]
    df = df.drop(columns=drop_cols)

    # Fix a data type: this is a category, not a true numeric measure
    if "MS_SubClass" in df.columns:
        df["MS_SubClass"] = df["MS_SubClass"].astype(str)

    # Fill some important numeric columns with the median
    numeric_fill = [
        "Lot_Frontage", "Mas_Vnr_Area", "BsmtFin_SF_1", "BsmtFin_SF_2",
        "Bsmt_Unf_SF", "Total_Bsmt_SF", "Garage_Area", "Garage_Cars", "Garage_Yr_Blt"
    ]
    for col in numeric_fill:
        if col in df.columns:
            df[col] = df[col].fillna(df[col].median())

    # Fill text columns with 'None'
    for col in df.select_dtypes(include="object").columns:
        df[col] = df[col].fillna("None")

    # Fill any remaining numeric nulls with the median
    for col in df.select_dtypes(include=np.number).columns:
        if df[col].isnull().sum() > 0:
            df[col] = df[col].fillna(df[col].median())

    # Remove duplicates
    df = df.drop_duplicates()

    # Cap extreme target values at the 99th percentile
    cap_value = df["SalePrice"].quantile(0.99)
    df["SalePrice"] = df["SalePrice"].clip(upper=cap_value)

    return df
```

```python
clean_df = clean_data(df)

print("Clean shape:", clean_df.shape)
print("Remaining missing values:", int(clean_df.isnull().sum().sum()))
print("SalePrice 99th percentile cap:", round(clean_df["SalePrice"].quantile(0.99), 2))
```

```python
# Final checks
assert clean_df["SalePrice"].isnull().sum() == 0
assert (clean_df["SalePrice"] > 0).all()
assert clean_df.shape[1] == 77

print("All checks passed.")
```

```python
# Save cleaned data
clean_df.to_csv("data/cleaned/ames_cleaned.csv", index=False)
clean_df.head()
```

## Cleaning summary in simple words
- I dropped a few columns with too many missing values.
- I filled numeric missing values with the median.
- I filled text missing values with `"None"`.
- I converted `MS_SubClass` to text because it behaves like a category.
- I capped very high `SalePrice` values instead of deleting them.
- I saved the cleaned file for the next notebook.
