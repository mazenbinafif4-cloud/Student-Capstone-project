# Phase 3 - EDA
This notebook focuses on simple charts and short insights.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/cleaned/ames_features.csv")
df.head()
```

```python
# Histograms for 3 numeric features
num_cols = ["SalePrice", "Gr_Liv_Area", "Total_Bsmt_SF"]

for col in num_cols:
    plt.figure(figsize=(7, 4))
    plt.hist(df[col].dropna(), bins=30)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

### Quick histogram insights
- `SalePrice` is right-skewed, which is common for housing prices.
- `Gr_Liv_Area` also has a long right tail.
- `Total_Bsmt_SF` shows many homes in the middle range and fewer very large basements.

```python
# Grouped boxplots
plt.figure(figsize=(9, 4))
sns.boxplot(data=df, x="Overall_Qual", y="SalePrice")
plt.title("SalePrice by Overall Quality")
plt.show()

top_neighborhoods = df["Neighborhood"].value_counts().head(8).index
plt.figure(figsize=(11, 4))
sns.boxplot(data=df[df["Neighborhood"].isin(top_neighborhoods)], x="Neighborhood", y="SalePrice")
plt.xticks(rotation=45)
plt.title("SalePrice by Neighborhood (Top 8 by count)")
plt.show()
```

### Boxplot insights
- Houses with better quality ratings are usually much more expensive.
- Price also changes a lot by neighborhood.

```python
# Correlation heatmap: top 10 numeric features with the target
top10 = (
    df.select_dtypes(include=np.number)
      .corr()["SalePrice"]
      .abs()
      .sort_values(ascending=False)
      .head(10)
      .index
)

plt.figure(figsize=(8, 6))
sns.heatmap(df[top10].corr(), cmap="Blues")
plt.title("Top 10 Features Correlated with SalePrice")
plt.show()
```

### Heatmap insight
- `Overall_Qual` is one of the strongest features connected to `SalePrice`.
- `Gr_Liv_Area`, garage size, and basement size are also important.

```python
# Scatter plot with a third variable
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x="Gr_Liv_Area",
    y="SalePrice",
    hue="Overall_Qual",
    alpha=0.6
)
plt.title("Living Area vs SalePrice with Overall Quality")
plt.show()
```

### Scatter plot insight
- Bigger houses usually cost more.
- The trend is positive, but it starts to flatten a little for very large homes.
- Houses with higher quality often sit higher on the chart.

```python
# Groupby summary
neighborhood_summary = (
    df.groupby("Neighborhood")["SalePrice"]
      .mean()
      .sort_values(ascending=False)
)

neighborhood_summary.head(10)
```

### Groupby insight
- The most expensive neighborhoods include **NoRidge** and **NridgHt**.
- One of the lowest average-price neighborhoods is **MeadowV**.
- This tells me that location matters a lot in this dataset.
